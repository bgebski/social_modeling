```python
import random
import math
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow
from numpy.polynomial.polynomial import polyfit
```

```python
czas10 = []
czas50 = []
czas100 = []
czas200 = []
#inicjalizacja tablic przechowujących liczbę kroków każdej próby dla każdego rozmiaru
```

```python
counter = 0
while counter < 10:
    l = 10
    yesCount = 0 #inicjalizacja licznika 'jedynek'
    tab =[] #inicjazlizacja matrycy
    for x in range (0,l*l):
        tab.append(int(round(random.random(), 0))) #wylosowanie opinii dla każdego
    for x in tab:
        if x==1:
            yesCount = yesCount + 1
    liczbaKrokow = 0
    yesRecord = []
    yesRecord.append(yesCount)
    while yesCount < l*l and yesCount > 0:
        for x in range(0,l*l):
            #random in range 0,1,2,3 = czyli góra/dół itd
            direction = random.randrange(0, 4)
            if direction == 0:
                #gora
                if tab[x]==0 and tab[x-l]==1:
                    tab[x] = 1
                    yesCount = yesCount + 1
                elif tab[x]==1 and tab[x-l]==0:
                    tab[x] = 0
                    yesCount = yesCount - 1
            elif direction == 1:
                #prawo
                if x%l==l-1:
                    if tab[x]==0 and tab[x-l+1]==1:
                        tab[x] = 1
                        yesCount = yesCount + 1
                    elif tab[x]==1 and tab[x-l+1]==0:
                        tab[x] = 0
                        yesCount = yesCount - 1
                else:
                    if tab[x]==0 and tab[x+1]==1:
                        tab[x] = 1
                        yesCount = yesCount + 1
                    elif tab[x]==1 and tab[x+1]==0:
                        tab[x] = 0
                        yesCount = yesCount - 1
            elif direction == 2:
                #dol
                if x>=l*(l-1):
                    if tab[x]==0 and tab[x-l*(l-1)]==1:
                        tab[x] = 1
                        yesCount = yesCount + 1
                    elif tab[x]==1 and tab[x-l*(l-1)]==0:
                        tab[x] = 0
                        yesCount = yesCount - 1
                else:
                    if tab[x]==0 and tab[x+l]==1:
                        tab[x] = 1
                        yesCount = yesCount + 1
                    elif tab[x]==1 and tab[x+l]==0:
                        tab[x] = 0
                        yesCount = yesCount - 1
            elif direction == 3:
                #lewo
                if x%l==0:
                    if tab[x]==0 and tab[x+l-1]==1:
                        tab[x] = 1
                        yesCount = yesCount + 1
                    elif tab[x]==1 and tab[x+l-1]==0:
                        tab[x] = 0
                        yesCount = yesCount - 1
                else:
                    if tab[x]==0 and tab[x-1]==1:
                        tab[x] = 1
                        yesCount = yesCount + 1
                    elif tab[x]==1 and tab[x-1]==0:
                        tab[x] = 0
                        yesCount = yesCount - 1
        liczbaKrokow = liczbaKrokow + 1
        yesRecord.append(yesCount)
        print("Ulamek osob YES: ",yesCount,"/",l*l)
        print("Liczba krokow:",liczbaKrokow)
        #wykonywanie "screenshotów", interwały dobrane arbitralnie
        if liczbaKrokow %6==1:
            name = str(l)+"_"+str(liczbaKrokow)
            dwad_array = np.reshape(tab,(l,l))
            plt.imshow(dwad_array)
            plt.colorbar()
            plt.savefig(name+'.png')
            plt.close()
    czas10.append(liczbaKrokow)
    print("KONIEC ITERACJI NR", counter)
    counter = counter +1
    ypoints = np.array(yesRecord)
    plt.plot(ypoints, linestyle = 'dotted')
    plt.show() 
    name = str(l)+"_time_"+str(counter)
    plt.savefig(name+'.png')
    plt.close()
print("KONIEC")
```
